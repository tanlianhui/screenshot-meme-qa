```python
conn = sqlite3.connect("scene_caption_correspondence.db")
cursor = conn.cursor()
cursor.execute("SELECT file_path, caption FROM images WHERE video_id = ?", ("1",))
images = cursor.fetchall()

# Display the images related to the video
for img in images:
    print(f"File: {img[0]}, Caption: {img[1]}")
```
